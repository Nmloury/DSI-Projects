### 1. Use LDA to find what topics are discussed on travel.stackexchange.com.

---

Text can be found in the posts and the comments datasets. The `ParentId` column in the posts dataset indicates what the "question" post was for a given post. Comment text can be merged onto the post they are part of with the `PostId` field.

The text may have some HTML tags. BeautifulSoup has convenient ways to get rid of markup or extract text if you need to. You can also parse the strings yourself if you like.

The tags dataset has the "tags" that the users have officially given the post.

**1.1 Implement LDA against the text features of the dataset(s).**

- This can be posts or a combination of posts and comments if you want more power.
- Find optimal **K/num_topics**.

**1.2 Compare your topics to the tags. Do the LDA topics make sense? How do they compare to the tags?**


```python
# Import Data
posts = pd.read_csv("./stack_exchange_travel/posts_train.csv")
comments = pd.read_csv("./stack_exchange_travel/comments_train.csv")
tags = pd.read_csv("./stack_exchange_travel/tags.csv")
users = pd.read_csv("./stack_exchange_travel/users.csv")
votes = pd.read_csv("./stack_exchange_travel/votes_train.csv")
```

```python
users.shape
```

```python
# See Shape
print tags.shape
print comments.shape
print posts.shape
```

```python
# Get text from posts df
def get_text(HTML):
    if type(HTML) == str:
        HTML_soup = BeautifulSoup(HTML, 'html.parser')
        return HTML_soup.text
    else:
        return None

posts['Text'] = posts['Body'].apply(get_text)
```

```python
# Drop posts with no text
posts.dropna(subset=['Text'], inplace=True)
```

```python
# Bag of Words on the Post Text
vectorizer = CountVectorizer(stop_words='english', max_df=.1, min_df=5)
X = vectorizer.fit_transform(posts.Text)
print X.shape
```

```python
# Create Vocab Dictionary
vocab = {v: k for k, v in vectorizer.vocabulary_.iteritems()}
```

```python
# Create LDA Model
lda = models.LdaModel(
    matutils.Sparse2Corpus(X, documents_columns=False),
    num_topics  =  3,
    passes      =  20,
    id2word     =  vocab
)
```

### 2. What makes an answer likely to be "accepted"?

---

**2.1 Build a model to predict whether a post will be marked as the answer.**

- This is a classification problem.
- You're free to use any of the machine learning algorithms or techniques we have learned in class to build the best model you can.
- NLP will be very useful here for pulling out useful and relevant features from the data. 
- Though not required, using bagging and boosting models like Random Forests and Gradient Boosted Trees will _probably_ get you the highest performance on the test data (but who knows!).


**2.2 Evaluate the performance of your classifier with a confusion matrix and accuracy. Explain how your model is performing.**

**2.3 Plot either a ROC curve or precision-recall curve (or both!) and explain what they tell you about your model.**

NOTE: You should only be predicting this for `PostTypeID=2` posts, which are the "answer" posts. This doesn't mean, however, that you can't or shouldn't use the parent questions as predictors!


```python
# Create Question and Answers dataframes
answers = posts[posts.PostTypeId == 2]
questions = posts.copy()
questions.columns = ['q_' + str(col) for col in questions.columns]
questions.head()

# Merge Answers with questions
answers = pd.merge(answers, questions, how='left', left_on='ParentId', right_on='q_Id')
print answers.shape
answers.head().T
```

```python
# Create Accepted column
answers['Accepted'] = answers['Id'] == answers['q_AcceptedAnswerId']
```

```python
# Drop Empty Columns
answers.dropna(axis=1, how='all', inplace=True)

# Find Relevant numeric columns
answers_num = answers.select_dtypes(exclude=['object'])
answers_num = answers_num[['CommentCount', 'Score', 'q_AnswerCount', 'q_CommentCount', 'q_FavoriteCount', 'q_Score', 'q_ViewCount']]

# Fill NAs
answers_num.loc[:,'q_FavoriteCount'].fillna(value=0, inplace=True)
print answers_num.shape
```

```python
len(tfidf.get_feature_names())
```

```python
# Create Features from Answer and Question Text
tfidf = TfidfVectorizer(stop_words='english', max_df=.1, min_df=10, max_features=100)
X_a = tfidf.fit_transform(answers.Text)
X_a = pd.DataFrame(data=X_a.todense(), columns=tfidf.get_feature_names())
X_q = tfidf.fit_transform(answers.q_Text)
X_q = pd.DataFrame(data=X_q.todense(), columns=tfidf.get_feature_names())
print X_a.shape
print X_q.shape
```

```python
# Create Target and predictor
y = answers['Accepted']
answer_pred = pd.concat([answers_num, X_a, X_q], axis=1)
X = np.array(answer_pred)
print y.shape, X.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import StratifiedKFold, train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Fit Random Forest Classifier
rfc = RandomForestClassifier()
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=.25)
folds = StratifiedKFold(y=y_train, n_folds=3, shuffle=True, random_state=10)

rfc_param = {
    'n_estimators': [10,50,100],
    'max_depth': [2,5,None],
    'max_features': ['auto', 'log2']
}

rfc_gs = GridSearchCV(estimator=rfc, param_grid=rfc_param, cv=folds, verbose=1, n_jobs=-1)
```

```python
rfc_gs.fit(X_train, y_train)
```

```python
print "Accuracy: ", rfc_gs.best_score_
print "Parameters: ", rfc_gs.best_params_
```

```python
# Look at Feature Importance
rfc_best = rfc_gs.best_estimator_
f_importance = pd.DataFrame(rfc_best.feature_importances_)
f_names = pd.DataFrame(answer_pred.columns)
f_importance = pd.concat([f_names, f_importance], axis=1)
f_importance.columns = ['Feature Name', 'Importance']
f_importance.sort(columns='Importance', ascending=False, inplace=True)
f_importance[:10]
```

```python
# Print Confusion Matrix and Classification Report
y_pred = rfc_gs.predict(X_test)
print confusion_matrix(y_test, y_pred)
print classification_report(y_test, y_pred)
```

### 3. What is the score of a post?

---

**3.1 Build a model that predicts the score of a post.**

- This is a regression problem now. 
- You can and should be predicting score for both "question" and "answer" posts, so keep them both in your dataset.
- Again, use any techniques that you think will get you the best model.

**3.2 Evaluate the performance of your model with cross-validation and report the results.**

**3.3 What is important for determining the score of a post, if anything?**


```python
posts.columns
```

```python
# Find Relevant Post numeric columns
posts_num = posts.select_dtypes(exclude=['object'])
posts_num = posts_num[['AnswerCount', 'CommentCount', 'FavoriteCount', 'PostTypeId', 'ViewCount']]

# Fill NAs
print posts_num.isnull().sum()
posts_num.fillna(value=0, inplace=True)
print posts_num.shape
posts_num.reset_index(drop=True, inplace=True)
```

```python
# Create Features from Posts Text
tfidf = TfidfVectorizer(stop_words='english', max_df=.1, min_df=10, max_features=100)
X_text = tfidf.fit_transform(posts.Text)
X_text = pd.DataFrame(data=X_text.todense(), columns=tfidf.get_feature_names())
print X_text.shape
X_text.reset_index(drop=True, inplace=True)
```

```python
# Create Target and Predictor
y = np.array(posts['Score'])
X = np.array(pd.concat([posts_num, X_text], axis=1))
print y.shape, X.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import KFold
```

```python
# Fit Random Forest Regressor
rfr = RandomForestRegressor()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25)
folds = KFold(n=y_train.shape[0], n_folds=3, shuffle=True, random_state=10)

rfr_param = {
    'n_estimators': [10,100,1000],
    'max_depth': [2,5,None],
    'max_features': ['auto', 'log2']
}

rfr_gs = GridSearchCV(estimator=rfr, param_grid=rfr_param, cv=3, verbose=1, n_jobs=-1)
```

```python
# Fit the Model
rfr_gs.fit(X_train, y_train)
```

```python
# Results
print "R^2: ", rfr_gs.best_score_
print "Parameters: ", rfr_gs.best_params_
```

```python
# Look at Feature Importance
rfr_best = rfr_gs.best_estimator_
f_importance = pd.DataFrame(rfr_best.feature_importances_)
f_names = pd.DataFrame(pd.concat([posts_num, X_text], axis=1).columns)
f_importance = pd.concat([f_names, f_importance], axis=1)
f_importance.columns = ['Feature Name', 'Importance']
f_importance.sort(columns='Importance', ascending=False, inplace=True)
f_importance[:10]
```

### 4. How many views does a post have?

---

**4.1 Build a model that predicts the number of views a post has.**

- This is another regression problem. 
- Predict the views for all posts, not just the "answer" posts.

**4.2 Evaluate the performance of your model with cross-validation and report the results.**

**4.3 What is important for the number of views a post has, if anything?**

```python
# Copy DF
posts_views = posts.copy()

# Drop Posts with Missing views
posts_views.dropna(subset=['ViewCount'], inplace=True)

# Find Relevant Post numeric columns
posts_views_num = posts_views.select_dtypes(exclude=['object'])
posts_views_num = posts_views_num[['AnswerCount', 'CommentCount', 'FavoriteCount', 'PostTypeId', 'Score']]

# Fill NAs
print posts_views_num.isnull().sum()
posts_views_num.fillna(value=0, inplace=True)
print posts_views_num.shape
posts_views_num.reset_index(drop=True, inplace=True)
```

```python
# Create Features from Posts Text
tfidf = TfidfVectorizer(stop_words='english', max_df=.1, min_df=10, max_features=100)
X_text = tfidf.fit_transform(posts_views.Text)
X_text = pd.DataFrame(data=X_text.todense(), columns=tfidf.get_feature_names())
print X_text.shape
X_text.reset_index(drop=True, inplace=True)
```

```python
# Create Target and Predictor
y = np.array(posts_views['ViewCount'])
X = np.array(pd.concat([posts_views_num, X_text], axis=1))
print y.shape, X.shape
```

```python
# Fit Random Forest Regressor
rfr = RandomForestRegressor()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25)
folds = KFold(n=y_train.shape[0], n_folds=3, shuffle=True, random_state=10)

rfr_param = {
    'n_estimators': [10,100,1000],
    'max_depth': [2,5,None],
    'max_features': ['auto', 'log2']
}

rfr_gs = GridSearchCV(estimator=rfr, param_grid=rfr_param, cv=folds, verbose=1, n_jobs=-1)
```

```python
# Fit the Model
rfr_gs.fit(X_train, y_train)
```

```python
# Results
print "R^2: ", rfr_gs.best_score_
print "Parameters: ", rfr_gs.best_params_
```

```python
# Look at Feature Importance
rfr_best = rfr_gs.best_estimator_
f_importance = pd.DataFrame(rfr_best.feature_importances_)
f_names = pd.DataFrame(pd.concat([posts_num, X_text], axis=1).columns)
f_importance = pd.concat([f_names, f_importance], axis=1)
f_importance.columns = ['Feature Name', 'Importance']
f_importance.sort(columns='Importance', ascending=False, inplace=True)
f_importance[:10]
```

```python
#PLot
y_pred = rfr_best.predict(X_test)
sb.regplot(x=np.log1p(y_pred), y=np.log1p(y_test))
plt.xlabel('Log Predicted Views')
plt.ylabel('Log Actual Views')
```
